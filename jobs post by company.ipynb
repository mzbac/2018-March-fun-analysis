```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('companyName_count.csv')
```

```python
df.head(10)
```

```python
labels = df["companyName"].values[0:11]
data = df["count"].values[0:11]

plt.figure(figsize=(120,60))
plt.bar(range(len(data)), data, tick_label=labels)
plt.tick_params(labelsize=40)
plt.show()
```
